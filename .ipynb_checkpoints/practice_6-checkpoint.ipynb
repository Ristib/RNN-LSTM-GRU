PyTorch

```python
# Импортирует необходимые бибилотеки

import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, roc_curve
from gensim.models import Word2Vec
from torch.utils.data import DataLoader, TensorDataset
import numpy as np

# Проверка на наличие CUDA
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Загрузка файла в Google Colab
from google.colab import drive
drive.mount("/content/drive")

train_file_path = "/content/drive/MyDrive/Colab Notebooks/Advansed/practice_6/FakeNewsNet.csv"

# Загрузка данных в DataFrame
data = pd.read_csv(train_file_path)

# Комбинирую текст из нужных столбцов и приводим к строковому типу
data['text'] = data['title'].astype(str) + ' ' + data['news_url'].astype(str) + ' ' + data['source_domain'].astype(str)

# Проверка и замена NaN на пустые строки
data['text'] = data['text'].fillna('')

# Определяю поля TEXT и LABEL
texts = data['text'].values
labels = data['real'].values

# Токенизация текстов
tokenized_texts = [text.split() for text in data['text']]

# Обучение модели Word2Vec
w2v_model = Word2Vec(sentences=tokenized_texts, vector_size=100, window=5, min_count=1, workers=4)
w2v_model.save('w2v_model.txt')

# Получаею векторы слов
word2vec_vectors = w2v_model.wv

# Преобразую тексты в последовательности индексов
max_len = 50
vocab = {word: i+1 for i, word in enumerate(word2vec_vectors.index_to_key)}
sequences = [[vocab.get(word, 0) for word in text.split()] for text in texts]
sequences = [seq[:max_len] + [0] * (max_len - len(seq)) if len(seq) < max_len else seq[:max_len] for seq in sequences]

# Разделяю данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(sequences, labels, test_size=0.2, random_state=42)

# Преобразую данные в тензоры
X_train_tensor = torch.LongTensor(X_train)
y_train_tensor = torch.FloatTensor(y_train)
X_test_tensor = torch.LongTensor(X_test)
y_test_tensor = torch.FloatTensor(y_test)

# Создаю датасеты и загрузчики
train_data = TensorDataset(X_train_tensor, y_train_tensor)
test_data = TensorDataset(X_test_tensor, y_test_tensor)

train_iterator = DataLoader(train_data, batch_size=64, shuffle=True)
test_iterator = DataLoader(test_data, batch_size=64)

# Общая архитектура для RNN, LSTM и GRU
class RNNModel(nn.Module):
    def __init__(self, input_dim, embedding_dim, hidden_dim, output_dim, n_layers, bidirectional, dropout, rnn_type='LSTM'):
        super().__init__()

        # Эмбеддинг слой
        self.embedding = nn.Embedding(input_dim, embedding_dim)

        # Выбор типа RNN слоя (RNN, LSTM, GRU)
        self.rnn_type = rnn_type  # добавляю тип слоя в объект

        if rnn_type == 'LSTM':
            self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers,
                               bidirectional=bidirectional, dropout=dropout, batch_first=True)
        elif rnn_type == 'GRU':
            self.rnn = nn.GRU(embedding_dim, hidden_dim, num_layers=n_layers,
                              bidirectional=bidirectional, dropout=dropout, batch_first=True)
        else:
            self.rnn = nn.RNN(embedding_dim, hidden_dim, num_layers=n_layers,
                              bidirectional=bidirectional, dropout=dropout, batch_first=True)

        # Полносвязный слой
        fc_input_dim = hidden_dim * 2 if bidirectional else hidden_dim
        self.fc = nn.Linear(fc_input_dim, output_dim)

        # Dropout
        self.dropout = nn.Dropout(dropout)

    def forward(self, text):
        # Текст -> эмбеддинги
        embedded = self.embedding(text)

        # Подаю эмбеддинги в RNN
        if self.rnn_type == 'LSTM':
            output, (hidden, cell) = self.rnn(embedded)
        elif self.rnn_type == 'GRU':
            output, hidden = self.rnn(embedded)
        else:  # Для обычного RNN возвращаю только output и hidden
            output, hidden = self.rnn(embedded)

        # Если модель двунаправленная, объединяю последние слои вперед-назад
        if self.rnn.bidirectional:
            hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        else:
            hidden = hidden[-1,:,:]

        # Применяю dropout и передаю в полносвязный слой
        final_output = self.fc(self.dropout(hidden))

        # Применяю функцию активации sigmoid для получения значений в диапазоне [0, 1]
        return torch.sigmoid(final_output).squeeze(1)

# Параметры модели
INPUT_DIM = len(vocab) + 1
EMBEDDING_DIM = 100
HIDDEN_DIM = 64
OUTPUT_DIM = 1
N_LAYERS = 2
BIDIRECTIONAL = True
DROPOUT = 0.8

# Получаю вектора Word2Vec
pretrained_embeddings = np.zeros((len(vocab) + 1, 100))  # len(vocab) + 1 для PAD токена
for word, idx in vocab.items():
    if word in word2vec_vectors:
        pretrained_embeddings[idx] = word2vec_vectors[word]
    else:
        pretrained_embeddings[idx] = np.random.normal(scale=0.6, size=(100,))  # случайные вектора для неизвестных слов

# Преобразую в тензор
pretrained_embeddings = torch.FloatTensor(pretrained_embeddings)

# Функции для обучения и оценки
def train(model, iterator, optimizer, criterion):
    model.train()
    epoch_loss = 0
    epoch_acc = 0
    for texts, labels in iterator:
        texts, labels = texts.to(device), labels.to(device).float()
        optimizer.zero_grad()
        # Получаю предсказания из модели
        predictions = model(texts).squeeze()  # Просто .squeeze() для удаления одиночных размерностей

        # Проверяю, что predictions и labels имеют одинаковую форму
        assert predictions.shape == labels.shape, f"Shape mismatch: {predictions.shape} vs {labels.shape}"

        loss = criterion(predictions, labels)
        acc = binary_accuracy(predictions, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        epoch_acc += acc.item()
    return epoch_loss / len(iterator), epoch_acc / len(iterator)

def evaluate(model, iterator, criterion):
    model.eval()
    epoch_loss = 0
    epoch_acc = 0
    with torch.no_grad():
        for texts, labels in iterator:
            texts, labels = texts.to(device), labels.to(device).float()
            predictions = model(texts).squeeze()
            loss = criterion(predictions, labels)
            acc = binary_accuracy(predictions, labels)
            epoch_loss += loss.item()
            epoch_acc += acc.item()
    return epoch_loss / len(iterator), epoch_acc / len(iterator)

# Функция для расчёта точности
def binary_accuracy(preds, y):
    rounded_preds = torch.round(preds)
    correct = (rounded_preds == y).float()
    acc = correct.sum() / len(correct)
    return acc

# Получение предсказаний
def get_predictions(model, iterator):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for texts, labels in iterator:
            texts, labels = texts.to(device), labels.to(device).float()
            predictions = model(texts).squeeze()
            all_preds.append(predictions.cpu())
            all_labels.append(labels.cpu())
    return torch.cat(all_preds), torch.cat(all_labels)

# Обучение модели для каждой архитектуры (RNN, LSTM, GRU)
def train_and_evaluate(rnn_type):
    print(f"Обучаем модель {rnn_type}...")

    # Инициализация модели
    model = RNNModel(INPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM, OUTPUT_DIM, N_LAYERS, BIDIRECTIONAL, DROPOUT, rnn_type).to(device)

    # Копирую предварительно обученные векторы в слой эмбеддингов
    model.embedding.weight.data.copy_(pretrained_embeddings)

    # Оптимизатор и функция потерь
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    criterion = nn.BCELoss()

    # Обучение модели (10 эпох)
    N_EPOCHS = 5
    train_losses, test_losses, train_accs, test_accs = [], [], [], []

    for epoch in range(N_EPOCHS):
        train_loss, train_acc = train(model, train_iterator, optimizer, criterion)
        test_loss, test_acc = evaluate(model, test_iterator, criterion)
        train_losses.append(train_loss)
        test_losses.append(test_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        print(f'Эпоха {epoch+1}: Потеря на обучении: {train_loss:.4f}, Точность на обучении: {train_acc:.4f}, Потеря на тестировании: {test_loss:.4f}, Точность на тестировании: {test_acc:.4f}')

    # Построение графиков потерь и точности
    plt.figure(figsize=(10,5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(test_losses, label='Test Loss')
    plt.title(f'График потерь для {rnn_type}')
    plt.xlabel('Эпоха')
    plt.ylabel('Потери')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,5))
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(test_accs, label='Test Accuracy')
    plt.title(f'График точности для {rnn_type}')
    plt.xlabel('Эпоха')
    plt.ylabel('Точность')
    plt.legend()
    plt.show()

    # Получение предсказаний
    preds, labels = get_predictions(model, test_iterator)
    rounded_preds = torch.round(preds)

    # Рассчёт метрик
    accuracy = accuracy_score(labels, rounded_preds)
    f1 = f1_score(labels, rounded_preds)
    roc_auc = roc_auc_score(labels, preds)

    print(f'Результаты {rnn_type}:')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'F1-Score: {f1:.4f}')
    print(f'ROC AUC: {roc_auc:.4f}')

    # Построение ROC AUC
    fpr, tpr, thresholds = roc_curve(labels, preds)
    plt.figure(figsize=(10,5))
    plt.plot(fpr, tpr, color='blue', label=f'ROC AUC = {roc_auc:.4f}')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.title(f'ROC AUC Curve для {rnn_type}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()

# Обучаю и оцениваю каждую модель
for rnn_type in ['RNN', 'LSTM', 'GRU']:
    train_and_evaluate(rnn_type)
```

TensorFlow

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, roc_curve
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
import nltk

# Загрузка данных
nltk.download('stopwords')
nltk.download('punkt')

# Загрузка файла (предполагается Google Colab)
from google.colab import drive
drive.mount("/content/drive")

train_file_path = "/content/drive/MyDrive/Colab Notebooks/Advansed/practice_6/FakeNewsNet.csv"

# Загрузка данных в DataFrame
data = pd.read_csv(train_file_path)

# Предобработка данных
data = data[['title', 'news_url', 'source_domain', 'real']].dropna()
data['text'] = data['title'] + ' ' + data['news_url'] + ' ' + data['source_domain']
data['text'] = data['text'].str.lower().apply(lambda x: re.sub(r'[^a-z\s]', '', x))

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

data['text'] = data['text'].apply(preprocess_text)

# Токенизация текста
max_size = 10000  # Размер словаря
tokenizer = Tokenizer(num_words=max_size)
tokenizer.fit_on_texts(data['text'].values)
sequences = tokenizer.texts_to_sequences(data['text'].values)

# Дополнение последовательностей
max_len = 50
X = pad_sequences(sequences, maxlen=max_len)

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, data['real'], test_size=0.2, random_state=42)

# Параметры модели
vocab_size = max_size
embedding_dim = 50  # Уменьшение размерности эмбеддингов
sequence_length = X_train.shape[1]

# Функция для построения моделей (RNN, LSTM, GRU)
def build_model(model_type):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_length,
                        embeddings_regularizer=l2(0.001)))

    if model_type == 'RNN':
        model.add(SimpleRNN(16, return_sequences=False, kernel_regularizer=l2(0.001)))
    elif model_type == 'LSTM':
        model.add(LSTM(16, return_sequences=False, kernel_regularizer=l2(0.001)))
    elif model_type == 'GRU':
        model.add(GRU(16, return_sequences=False, kernel_regularizer=l2(0.001)))

    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid', kernel_regularizer=l2(0.001)))
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])

    return model

# Early stopping для предотвращения переобучения
early_stopping = EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

# Обучение и оценка каждой модели
def train_and_evaluate(model_type):
    model = build_model(model_type)
    print(f"Обучение модели {model_type}...")

    history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test),
                        callbacks=[early_stopping], batch_size=64, verbose=1)

    # Оценка модели
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    print(f'{model_type} Test Accuracy: {acc:.4f}')

    return model, history

# Функция для построения графиков
def plot_metrics(history, title):
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Test Loss')
    plt.title(f'Train and Test Loss ({title})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Test Accuracy')
    plt.title(f'Train and Test Accuracy ({title})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

# ROC-AUC
def plot_roc_auc(model, X_test, y_test):
    preds = model.predict(X_test)
    roc_auc = roc_auc_score(y_test, preds)
    fpr, tpr, _ = roc_curve(y_test, preds)
    plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.4f}')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC AUC Curve')
    plt.legend()
    plt.show()

# Оценка модели
def evaluate_model(model, X_test, y_test):
    preds = (model.predict(X_test) > 0.5).astype("int32")
    accuracy = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    roc_auc = roc_auc_score(y_test, model.predict(X_test))

    print(f'Accuracy: {accuracy:.4f}')
    print(f'F1-Score: {f1:.4f}')
    print(f'ROC AUC: {roc_auc:.4f}')

# 1. Обучение и оценка RNN
rnn_model, history_rnn = train_and_evaluate('RNN')
plot_metrics(history_rnn, 'RNN')
plot_roc_auc(rnn_model, X_test, y_test)
evaluate_model(rnn_model, X_test, y_test)

# 2. Обучение и оценка LSTM
lstm_model, history_lstm = train_and_evaluate('LSTM')
plot_metrics(history_lstm, 'LSTM')
plot_roc_auc(lstm_model, X_test, y_test)
evaluate_model(lstm_model, X_test, y_test)

# 3. Обучение и оценка GRU
gru_model, history_gru = train_and_evaluate('GRU')
plot_metrics(history_gru, 'GRU')
plot_roc_auc(gru_model, X_test, y_test)
evaluate_model(gru_model, X_test, y_test)
```
